```python
# Perfect sensor
def sense(x):
    return x
```

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

```python
def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    ux = -sin(t)
    uy = cos(t)
    uz = 2/(5*pi)
    return array([ux, uy, uz])
```

```python
tf = 5*pi
Δt = 0.01    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([1., 0., 0.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```

```python
fig = figure()
ax = axes(projection='3d')
ax.plot3D(x_log[:,0], x_log[:,1], x_log[:,2])
```
