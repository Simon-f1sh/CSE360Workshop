```python
%pylab inline
plt.style.use('seaborn-whitegrid')
pylab.rcParams['figure.figsize'] = (10.0, 6.0)
```

```python
# Obstacles
def plot_cir(c):
    axis = gca()
    axis.add_patch(Circle(
        c[0],
        c[1],
        color=c[2]
    ))
    
def plot_free_space(c):
    axis = gca()
    axis.add_patch(Circle(
        c[0],
        c[1]+0.6,
        color='gray'
    ))
```

```python
# Obstacles info
r1 = sqrt(square(4) + square(1))/2
r2 = 1/2

cu0 = ((2.725, -0.00000027), r1, 'k')
cu1 = ((7.425, 4.725), r1, 'k')
cu3 = ((-1.025, -1.5), r1, 'k')
cu4 = ((-6.7, 6.4), r1, 'k')
cu6 = ((-5.15, -6.2), r1, 'k')

cy5 = ((2.425, 4.075), r2, 'b')
cy6 = ((4.4, 0.875), r2, 'b')
cy7 = ((2.825, -4.85), r2, 'b')
cy8 = ((-0.525, 6.3), r2, 'b')
cy9 = ((-5.125, 3.425), r2, 'b')
cy10 = ((-1.775, 2.1), r2, 'b')
cy11 = ((-5.8, -0.975), r2, 'b')
cy12 = ((7.275, -7.75), r2, 'b')

goal = ((-7.425, -7.925), 0.75, 'green')
```

```python
# Plot
gca().set_aspect("equal")
# ylim((-9,14))
# xlim((-7,12))

plot_free_space(cu0)
plot_free_space(cu1)
plot_free_space(cu3)
plot_free_space(cu4)
plot_free_space(cu6)

plot_free_space(cy5)
plot_free_space(cy6)
plot_free_space(cy7)
plot_free_space(cy8)
plot_free_space(cy9)
plot_free_space(cy10)
plot_free_space(cy11)
plot_free_space(cy12)

plot_cir(cu0)
plot_cir(cu1)
plot_cir(cu3)
plot_cir(cu4)
plot_cir(cu6)

plot_cir(cy5)
plot_cir(cy6)
plot_cir(cy7)
plot_cir(cy8)
plot_cir(cy9)
plot_cir(cy10)
plot_cir(cy11)
plot_cir(cy12)

plot_cir(goal)

plot()
```
